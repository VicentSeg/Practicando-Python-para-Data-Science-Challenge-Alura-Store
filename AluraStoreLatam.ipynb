### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



En este primer análisis, se calcula el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
import matplotlib.pyplot as plt

# Ingresos de cada tienda
ingresos_numericos = [
    tienda["Precio"].sum(),
    tienda2["Precio"].sum(),
    tienda3["Precio"].sum(),
    tienda4["Precio"].sum()
]

# Nombres de las tiendas
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# Crear el gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(tiendas, ingresos_numericos, color='skyblue')
plt.title("Ingreso total por tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingreso ($)")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar los valores sobre cada barra
for i, v in enumerate(ingresos_numericos):
    plt.text(i, v + max(ingresos_numericos)*0.01, f"${v:,.3f}", ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

vamos a calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
# Lista de dataframes y nombres para recorrer en bucle
tiendas_df = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
# Recorremos cada tienda para analizar por categoría
for i, df in enumerate(tiendas_df):
    print(f"\n🏷️ Análisis por categoría - {nombres_tiendas[i]}")

    # Contar cuántos productos se vendieron por categoría
    ventas_por_categoria = df["Categoría del Producto"].value_counts()

    # Categoría más popular
    categoria_top = ventas_por_categoria.idxmax()
    ventas_top = ventas_por_categoria.max()

    print(f"Categoría más vendida: {categoria_top} ({ventas_top} ventas)")

    # Mostrar tabla de todas las categorías
    print("\nVentas por categoría:")
    print(ventas_por_categoria)

    # Gráfico: Top 5 categorías más vendidas
    plt.figure(figsize=(8, 4))
    ventas_por_categoria.head(5).plot(kind='bar', color='mediumseagreen')
    plt.title(f"Top 5 Categorías más vendidas - {nombres_tiendas[i]}")
    plt.xlabel("Categoría")
    plt.ylabel("Cantidad de Ventas")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

# 3. Calificación promedio de la tienda


se calculara las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
# Calcular promedio de calificación por tienda
calificaciones = {
    "Tienda": nombres_tiendas,
    "Calificación Promedio": [
        tienda["Calificación"].mean(),
        tienda2["Calificación"].mean(),
        tienda3["Calificación"].mean(),
        tienda4["Calificación"].mean()
    ]
}

# Crear DataFrame con los resultados
df_calificaciones = pd.DataFrame(calificaciones)

# Redondear
df_calificaciones["Calificación Promedio"] = df_calificaciones["Calificación Promedio"].round(3)

# Mostrar tabla
print(df_calificaciones)
```

```python
plt.figure(figsize=(8, 4))
plt.bar(df_calificaciones["Tienda"], df_calificaciones["Calificación Promedio"], color='plum')
plt.title("Calificación Promedio por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Calificación Promedio")
plt.ylim(0, 5)  # Asumiendo que la calificación va de 1 a 5
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar valores sobre las barras
for i, v in enumerate(df_calificaciones["Calificación Promedio"]):
    plt.text(i, v + 0.05, f"{v:.3f}", ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

En este primer análisis, se calcula el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python

# Recorrer cada tienda
for i, df in enumerate(tiendas_df):
    print(f"\n🔎 {nombres_tiendas[i]}")

    # Contar ventas por producto
    ventas_por_producto = df["Producto"].value_counts()

    # Producto más vendido
    producto_mas = ventas_por_producto.idxmax()
    ventas_mas = ventas_por_producto.max()

    # Producto menos vendido
    producto_menos = ventas_por_producto.idxmin()
    ventas_menos = ventas_por_producto.min()

    print(f"Producto más vendido: {producto_mas} ({ventas_mas} ventas)")
    print(f"Producto menos vendido: {producto_menos} ({ventas_menos} venta)")

    # Gráfico: Top 5 productos más vendidos
    plt.figure(figsize=(8, 4))
    ventas_por_producto.head(5).plot(kind='bar', color='skyblue')
    plt.title(f"Top 5 Productos más vendidos - {nombres_tiendas[i]}")
    plt.xlabel("Producto")
    plt.ylabel("Cantidad de Ventas")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

# 5. Envío promedio por tienda

Se calculara el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

```python
# Calcular costo de envío promedio por tienda
costos_envio = {
    "Tienda": nombres_tiendas,
    "Costo de Envío Promedio": [
        tienda["Costo de envío"].mean(),
        tienda2["Costo de envío"].mean(),
        tienda3["Costo de envío"].mean(),
        tienda4["Costo de envío"].mean()
    ]
}

# Crear DataFrame con los resultados
df_costos_envio = pd.DataFrame(costos_envio)

plt.figure(figsize=(8, 4))
plt.bar(df_costos_envio["Tienda"], df_costos_envio["Costo de Envío Promedio"], color='lightcoral')
plt.title("Costo de Envío Promedio por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Costo de Envío Promedio ($)")
plt.ylim(0, df_costos_envio["Costo de Envío Promedio"].max() + 5)  # Ajustar el límite superior
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar valores sobre las barras
for i, v in enumerate(df_costos_envio["Costo de Envío Promedio"]):
    plt.text(i, v + 0.05, f"${v:.3f}", ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```
